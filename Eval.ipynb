```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

df = pd.read_csv('Results/Finetune/Results.csv')
```

```python
# Filter data for 'finetune' models
subset_finetune = df[df["Model"] == "finetune"]

# Create a figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Plot lineplots for 'finetune' models
sns.lineplot(data=subset_finetune, x="Percentage", y="MSE", label="MSE (finetune)", ax=ax)
sns.lineplot(data=subset_finetune, x="Percentage", y="MAE", label="MAE (finetune)", ax=ax)
sns.lineplot(data=subset_finetune, x="Percentage", y="RMSE", label="RMSE (finetune)", ax=ax)

# Filter data for 'scratch' models
#subset_scratch = df[df["Model"] == "scratch"]

# Plot pointplots for 'scratch' models
#sns.lineplot(data=subset_scratch, x="Percentage", y="MSE", label="MSE (scratch)", ax=ax, color='blue', marker='o')
#sns.lineplot(data=subset_scratch, x="Percentage", y="MAE", label="MAE (scratch)", ax=ax, color='orange', marker='o')
#sns.lineplot(data=subset_scratch, x="Percentage", y="RMSE", label="RMSE (scratch)", ax=ax, color='green', marker='o')

#plt.xlim(0, 50)
plt.ylabel("Error")
plt.title("Comparison of models trained on % of data")
plt.legend()
plt.show()

```

```python
# Create a table of MSE, MAE and RMSE for each model with columns for each percentage

summary_table = df.groupby(['Model','Percentage']).agg({'MSE': ['mean', 'std'],
                                         'MAE': ['mean', 'std'],
                                         'RMSE': ['mean', 'std']}).round(2)

summary_table.columns = ['MSE mean', 'MSE std', 'MAE mean', 'MAE std', 'RMSE mean', 'RMSE std']

summary_table['MSE'] = summary_table['MSE mean'].astype(str) + ' (' + summary_table['MSE std'].astype(str) + ')'
summary_table['MAE'] = summary_table['MAE mean'].astype(str) + ' (' + summary_table['MAE std'].astype(str) + ')'
summary_table['RMSE'] = summary_table['RMSE mean'].astype(str) + ' (' + summary_table['RMSE std'].astype(str) + ')'

summary_table = summary_table.drop(['MSE mean', 'MSE std', 'MAE mean', 'MAE std', 'RMSE mean', 'RMSE std'], axis=1)

summary_table
```
